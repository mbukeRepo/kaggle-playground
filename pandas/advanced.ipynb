```python
# merging two dataframes
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
```

```python
df1
```

```python
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
df1 = pd.DataFrame({"employee": ["Julia", "Marie", "Adam", "Nicole"], "department": ["Data Science", "Web Development", "Data Science", "Cyber Security"], "Year": ["2005", "2008", "2011", "2002"]})
```

```python
df1
```

```python
df2 = pd.DataFrame({"employee": ["Nicole", "Adam", "Julia", "Marie"], "country": ["Canada", "England", "USA", "Germany"], "salary": ["2200", "1600", "2000", "17500"]})
```

```python
df2
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on="employee", how="right")
```

```python
pd.merge(df1, df2, on="employee", how="left")
```

```python
df4 = pd.DataFrame({"employee": ["Julia", "Alex", "Adam", "Kayra"], "department": ["Data Science", "Web Development", "Data Science", "Cyber Security"]})
```

```python
df4
```

```python
df1
```

```python
pd.merge(df4, df1, on=["employee", "department"])
```

```python
pd.merge(df4, df1, on=["employee", "department"], how = "outer")
```

```python
pd.merge(df4, df1, on=["employee", "department"], how="inner")
```

```python
pd.merge(df4, df1, on=["employee", "department"], how="right")
```

```python
pd.merge(df4, df1, on=["employee", "department"], how="left")
```
